```python
import random
import time

# Interpolation Search Function
def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1
    probes = 0

    start_time = time.perf_counter()

    while low <= high and arr[low] <= key <= arr[high]:

        probes += 1

        # Avoid division by zero
        if arr[low] == arr[high]:
            if arr[low] == key:
                end_time = time.perf_counter()
                return low, probes, end_time - start_time
            break

        pos = low + int(
            ((key - arr[low]) * (high - low))
            / (arr[high] - arr[low])
        )

        if pos < 0 or pos >= len(arr):
            break

        if arr[pos] == key:
            end_time = time.perf_counter()
            return pos, probes, end_time - start_time

        elif arr[pos] < key:
            low = pos + 1

        else:
            high = pos - 1

    end_time = time.perf_counter()
    return -1, probes, end_time - start_time


# Manual Input Dataset
def manual_input():
    n = int(input("Enter number of elements: "))
    arr = list(map(int, input("Enter elements separated by space: ").split()))

    if len(arr) != n:
        print("Number of elements does not match!")
        return None

    return arr


# Random Uniform Dataset
def generate_uniform(n):
    start = random.randint(1, 20)
    gap = random.randint(2, 10)

    arr = [start + i * gap for i in range(n)]
    return arr


# Random Non-Uniform Dataset
def generate_non_uniform(n):
    arr = []
    value = random.randint(1, 10)

    for _ in range(n):
        value += random.randint(1, 50)
        arr.append(value)

    return arr


# Main Program
print("\nINTERPOLATION SEARCH ON USER-DEFINED DATASETS\n")

print("1. Uniform Dataset")
print("2. Non-Uniform Dataset")

dataset_choice = int(input("Choose dataset type: "))

print("\n1. Enter Data Manually")
print("2. Generate Random Data")

input_choice = int(input("Choose input method: "))

# Dataset Creation
if input_choice == 1:
    arr = manual_input()

else:
    n = int(input("Enter dataset size: "))

    if dataset_choice == 1:
        arr = generate_uniform(n)
    else:
        arr = generate_non_uniform(n)

    print("\nGenerated Dataset:")
    print(arr)

# Check Sorting
if arr != sorted(arr):
    sort_choice = input("\nDataset is not sorted. Sort before searching? (y/n): ")

    if sort_choice.lower() == 'y':
        arr.sort()
        print("Sorted Dataset:")
        print(arr)

# Search Key
key = int(input("\nEnter key to search: "))

# Perform Search
index, probes, exec_time = interpolation_search(arr, key)

# Results
print("\n----- SEARCH RESULT -----")

if index != -1:
    print(f"Key {key} found at index {index}")
else:
    print(f"Key {key} not found")

print(f"Number of probes/comparisons : {probes}")
print(f"Execution time : {exec_time:.10f} seconds")

# Performance Note
print("\n----- PERFORMANCE ANALYSIS -----")
if dataset_choice == 1:
    print("Uniform Dataset Selected")
    print("Interpolation Search performs efficiently on uniformly distributed data.")
else:
    print("Non-Uniform Dataset Selected")
    print("Interpolation Search may require more probes on non-uniform data and its performance can degrade toward O(n).")
```
